# Зависимости

```python
import math
import numpy
%matplotlib inline
import matplotlib.pyplot as pyplot

from common import *
```

# Параметры

```python
# Бит на символ
K = 2*3 # Должно быть кратно 2
# Кол-во символов QAM-M
M = (2**(K//2))**2

print "QAM-{0}".format(M)
print "Бит на символ = {0}".format(K)

# Амплитуда модуляции
A = 10.0

print "Амплитуда модуляции = {0}".format(A)

# Длина входного сигнала в символах
L = 16
# Длина входного сигнала в битах
N = K*L

print "Длина входного сигнала = {0}символов = {1}бит".format(L, N)

# Символьная частота/скорость входного сигнала
FS = 100*1000
TS = 1.0/FS

print "Символьная частота входного сигнала = {0}KГц".format(FS/1000)

# Длительность входного сигнала
DUR=L*TS
print "Длительность входного сигнала = {0}мс".format(DUR*1000)

# Частота несущей
F0 = 100*1000
T0 = 1.0/F0
W0 = 2.0*math.pi*F0

print "Частота несущей = {0}KГц".format(F0/1000)

# Кол-во отсчётов несущей на один символ входного сигнала
SCALE = 32

print "Кол-во отсчётов несущей на один символ = {0}".format(SCALE)

# Расчетная скорость
SPEED = K*FS

print "Расчетная скорость = {0}Kбит/сек".format(SPEED/1000)

# Шаблонный сигнал, посылаемый для настройки приёмника в самом начале передачи
TMPL_SYMB_0 = numpy.zeros(K//2)
TMPL_SYMB_1 = numpy.ones(K//2)
TMPL = numpy.concatenate([TMPL_SYMB_0, TMPL_SYMB_1, TMPL_SYMB_1, TMPL_SYMB_0, TMPL_SYMB_0, TMPL_SYMB_1, TMPL_SYMB_1, TMPL_SYMB_0])

print "Шаблонный сигнал:"
print TMPL
```

# Constellation Diagram

```python
def num_to_bit_seq(num, bit_count=K//2):
    return [(int(num) >> i) & 1 for i in xrange(bit_count)]
levels = map(lambda x: bit_seq_to_level(num_to_bit_seq(x), A), xrange(2**(K//2)))

for level in levels:
    pyplot.plot(levels, [level]*len(levels), "rx")
pyplot.title("Constellation Diagram")
pyplot.xlabel("I (synch)")
pyplot.ylabel("Q (quad)")
pyplot.xlim(-1.1*A, 1.1*A)
pyplot.ylim(-1.1*A, 1.1*A)
center_plot_axis()
pyplot.minorticks_on()
pyplot.grid()

pyplot.show()
```

# Входной сигнал

Cигнал состоит из шаблона и *рандомной* последовательности бит. Общая длина сигнала - *N бит*.

```python
# Входной сигнал
signal = numpy.concatenate([TMPL, numpy.random.random_integers(0, 1, N - len(TMPL))])

# График входного сигнала
pyplot.plot(xrange(len(signal)), signal, "ro:")
pyplot.title("Input signal (bits)")
pyplot.xlabel("Bit number")
pyplot.xlim(0, int(min(N, len(TMPL) + N//4)))
pyplot.ylim(-0.1, 1.1)
pyplot.show()
```

Первые K * len(TMPL) отсчётов - шаблон. Его перепады хорошо видны на графике. Шаблон подбирается таким образом, чтобы после модуляции максимизировать его мощность (вероятность прохождения по каналу) и упростить его детектироваие.

# Синфазная и квадратурная части входного сигнала

Это предварительный этап перед непосредственной модуляцией сигнала. Последовательность бит входного сигнала необходимо разбить на две части для каждого из каналов (синфазного и квадратурного).
Для этого:
1. Исходная последовательность бит разбивается на подпоследовательности по *K/2 бит*. То есть последовательность бит [0,1,0,1,1,0,1,1,1,0,0,0] будет представлена как [[0,1,0], [1,1,0], [1,1,1], [0,0,0]], при K = 6.
2. В *синфазную* часть *уходят* все четные подпоследовательности. В *квадратурную* - все нечетные. То есть *синфазная* часть [[0,1,0], [1,1,1]], и *квадратурную* - [[1,1,0], [0,0,0]].
3. Каждая подпоследовательность бит преобразуется в уровень *от -A до A* при помощи функции *bit_seq_to_level*.

Получаем две последовательности уровней *от -A до A* длиной *L* (кол-во символов).

```python
signal_parts = map(lambda x: bit_seq_to_level(x, A), signal.reshape(2*L, K//2))

# Синфазная часть
signal_synch_part = numpy.array(signal_parts[0:2*L:2])
# Квадратурная часть
signal_quad_part = numpy.array(signal_parts[1:2*L:2])

# Синфазная часть шаблонного сигнала
TMPL_SYNCH_PART = numpy.array(signal_synch_part[0:len(TMPL)//K])
TMPL_SYNCH_PART_PWR = sum(TMPL_SYNCH_PART*TMPL_SYNCH_PART)
# Квадратурная часть шаблонного сигнала
TMPL_QUAD_PART = numpy.array(signal_quad_part[0:len(TMPL)//K])
TMPL_QUAD_PART_PWR = sum(TMPL_QUAD_PART*TMPL_QUAD_PART)

# График синфазной части
pyplot.subplot(2, 1, 1)
t = numpy.linspace(0, DUR, L)
pyplot.plot(t*1000, signal_synch_part, "yo:")
pyplot.ylim(-1.5*A, 1.5*A)
pyplot.title("Input signal parts")
pyplot.ylabel("Synch part")
center_plot_axis()
pyplot.grid()

# График квадратурной части
pyplot.subplot(2, 1, 2)
pyplot.plot(t*1000, signal_quad_part, 'go:')
pyplot.ylim(-1.5*A, 1.5*A)
pyplot.xlabel("Time (ms)")
pyplot.ylabel("Quad part")
center_plot_axis()
pyplot.grid()

pyplot.show()
```

# Несущая

Данный график необходим для оценки адекватности выбранного параметра *SCALE* (кол-во отсчётов несущей на один символ). Слишком маленькие значения данного параметры приведут к потере точности при расчёте модулированного сигнала. Слишком большие - увеличивают требования к ресурсам (вычислительным или иным) приёмника/передатчика.

```python
t = numpy.linspace(0, DUR, SCALE*L)
# Несущая
carrier_sin = numpy.sin(W0*t)
carrier_cos = numpy.cos(W0*t)

# График несущей
pyplot.plot(t*1000, carrier_sin, "r.:")
pyplot.title("Carrier")
pyplot.xlabel("Time (ms)")
pyplot.xlim(0, 2*T0*1000)
pyplot.ylim(-1.1, 1.1)
center_plot_axis()

pyplot.show()
```

# Модулированный сигнал и его АЧХ

```python
# Модулированный сигнал
t = numpy.linspace(0, DUR, SCALE*L)
mod_signal = scale_signal(signal_synch_part, SCALE)*carrier_cos + scale_signal(signal_quad_part, SCALE)*carrier_sin

# График модулированного сигнала
pyplot.plot(t*1000, mod_signal, "r:")
pyplot.title("Mod-ed signal")
pyplot.xlabel("Time (ms)")
pyplot.ylim(-2*A - 0.1, 2*A + 0.1)
center_plot_axis()
pyplot.grid()

pyplot.show()

# АЧХ модулированного сигнала
(freq, afc) = compute_signal_afc(mod_signal, SCALE*L/DUR)
pyplot.plot(freq/1000, afc, "r-")
pyplot.title("AFC")
pyplot.xlabel("Freq (KHz)")
pyplot.grid()

pyplot.show()
```

# После прохождения сигнала через кабель

```python
import cable
```

```python
# Модулированный сигнал на выходе из кабеля
t = numpy.linspace(0, DUR, SCALE*L)
mod_signal_after_cable = cable.cable(mod_signal, noise_amp=A/2)

# График модулированного сигнала на выходе из кабеля
pyplot.plot(t*1000, mod_signal_after_cable, "r:")
pyplot.title("Mod-ed signal after cable")
pyplot.xlabel("Time (ms)")
k = 10**(cable.ATT/10)
pyplot.ylim((-2*A - 0.1)/k, (2*A + 0.1)/k)
center_plot_axis()
pyplot.grid()

pyplot.show()

# АЧХ модулированного сигнала на выходе из кабеля
(freq, afc) = compute_signal_afc(mod_signal_after_cable, SCALE*L/DUR)
pyplot.plot(freq/1000, afc, "r-")
pyplot.title("AFC")
pyplot.xlabel("Freq (KHz)")
pyplot.grid()

pyplot.show()
```

# Демодуляция и корректировка

Шаги работы приёмника:

1. Детектирование шаблона.
2. Демодуляция, путём умножения полченного сигнала на несущую. На *cos* для получения *синфазной* части, и на *sin* для получения *квадратурной* части
3. **TODO** Фильтр НЧ необходим, чтобы убрать "артефакты", полученные на предидущем шаге.
4. Так как приёмником некогерентен, следует произвести фазовый сдвиг, так чтобы максимизировать мощность шаблона. Шаблон должен подбираться так, чтобы любые отклонения фазы приёмника снижали его мощность после демодуляции.
5. В силу "затухания" в канале, производится корректировка уровней, исходя из, заранее известной, эталонной мощности *шаблона* и фактической мощности принятого шаблона.

```python
# Демодулированные части сигнала
demod_signal_synch_part = unscale_signal(2*mod_signal_after_cable*carrier_cos, SCALE)
demod_signal_quad_part = unscale_signal(2*mod_signal_after_cable*carrier_sin, SCALE)

# Демодулированная синфазная часть шаблонного сигнала
demod_tmpl_synch_part = demod_signal_synch_part[0:len(TMPL_SYNCH_PART)]
# Расчетное затухание 
tmpl_att = math.sqrt(TMPL_SYNCH_PART_PWR/sum(demod_tmpl_synch_part**2))

print "Расчетное затухание = {0}Дб".format(10*math.log(tmpl_att, 10))

demod_signal_synch_part = tmpl_att*demod_signal_synch_part
demod_signal_quad_part = tmpl_att*demod_signal_quad_part

# График синфазной части демодулированного сигнала
pyplot.subplot(2, 1, 1)
t = numpy.linspace(0, DUR, L)
pyplot.plot(t*1000, demod_signal_synch_part, "yo:")
pyplot.ylim(-1.5*A, 1.5*A)
pyplot.title("Output signal")
pyplot.ylabel("Synch part")
center_plot_axis()
pyplot.grid()

# График квадратурной части демодулированного сигнала
pyplot.subplot(2, 1, 2)
pyplot.plot(t*1000, demod_signal_quad_part, 'go:')
pyplot.ylim(-1.5*A, 1.5*A)
pyplot.xlabel("Time (ms)")
pyplot.ylabel("Quad part")
center_plot_axis()
pyplot.grid()

pyplot.show()
```

# Восстановленный сигнал

Демодулированные части сигнала преобразуется и "складываются" в битовую последовательность по обратному алгоритму

```python
restored_signal_synch_part = map(lambda x: level_to_bit_seq(x, K//2, A), demod_signal_synch_part)
restored_signal_quad_part = map(lambda x: level_to_bit_seq(x, K//2, A), demod_signal_quad_part)

restored_signal = numpy.concatenate(numpy.concatenate(zip(restored_signal_synch_part, restored_signal_quad_part)))

# График восстановленного сигнала
pyplot.plot(xrange(len(restored_signal)), restored_signal, "ro:")
pyplot.title("Restored signal")
pyplot.xlabel("Time (ms)")
pyplot.xlim(0, int(min(N, len(TMPL) + N//4)))
pyplot.ylim(-0.1, 1.1)
pyplot.show()
```
